```python
url = "https://ds-wheels.s3.us-east-1.amazonaws.com/m_4007307_sw_18_060_20220803.tif"
layer = COGTileLayer(data=url)
```

```python
view_state = {'longitude': -73.20972420038129,
 'latitude': 40.90307721701271,
 'zoom': 10.98785761041711}
```

```python
m = Map(layer, view_state=view_state)
```
