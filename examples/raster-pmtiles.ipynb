# PMTiles Raster example

This is a short example for using raster PMTiles sources in Lonboard.



## Dependencies

Install [`uv`](https://docs.astral.sh/uv) and then launch this notebook with:

```
uvx juv run examples/raster-pmtiles.ipynb
```

(The `uvx` command is included when installing `uv`).

## Imports

````python
from async_pmtiles import PMTilesReader
from obstore.store import HTTPStore

from lonboard import Map, RasterLayer
````

## Create PMTiles reader

Use the [`async-pmtiles`](https://github.com/developmentseed/async-pmtiles) library to create a new `PMTilesReader`.

Here we use [Obstore](https://developmentseed.org/obstore/latest/) to read data from a remote source. Obstore also lets you read data from AWS, Google Cloud Storage, Azure Storage, or local files.

Then we pass the path and store into `PMTilesReader.open` to instantiate a new reader.

````python
store = HTTPStore("https://air.mtn.tw")
reader = await PMTilesReader.open("flowers.pmtiles", store=store)
````

## Create RasterLayer from PMTiles source

Now we can create a new `RasterLayer` from this source, and display it on the map.

````python
layer = RasterLayer.from_pmtiles(reader)
````

````python
m = Map(layer)
m
````
