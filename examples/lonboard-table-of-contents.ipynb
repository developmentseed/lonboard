# Lonboard Table of Contents

This notebook demonstrates using the Lonboard controls to make a table of contents from a Lonboard map.

```python
import shapely
import geopandas as gpd

import lonboard
```

#### Make some simple dataframes and layers

This notebook is not to demonstrate the speed of Lonboard, rather, it is simply to show the table of contents.  Some simple geodataframes with one row each will be sufficient to make layers to add to a map.

```python
point = shapely.Point(-90, 38.8)
point_df = gpd.GeoDataFrame(
    [point],
    columns=["geometry"],
    geometry="geometry",
    crs=4326
)
point_layer = lonboard.ScatterplotLayer.from_geopandas(
    point_df,
    title="Point Layer",
    radius_min_pixels=10,
    get_fill_color=[255, 255, 0]
)

line = shapely.LineString([[-91, 38], [-90, 39]])
line_df = gpd.GeoDataFrame(
    [line],
    columns=["geometry"],
    geometry="geometry",
    crs=4326,
)
line_layer = lonboard.PathLayer.from_geopandas(
    line_df,
    title="Line Layer",
    width_min_pixels=10,
    get_color=[124, 124, 124]
)

polygon = shapely.Polygon([[-91, 38], [-90, 39], [-89, 38.5]])
polygon_df = gpd.GeoDataFrame(
    [polygon],
    columns=["geometry"],
    geometry="geometry",
    crs=4326
)
polygon_layer = lonboard.PolygonLayer.from_geopandas(
    polygon_df,
    title="Polygon Layer"
)
```

#### Make the map with the polygon and line layers and then make a table of contents from the map

The checkboxes in the table of contents toggle the layer's visibility.

If you click the settings gear for a layer in the TOC, the layer's traits are exposed and can be changed by the user

```python
lonboard_map = lonboard.Map([polygon_layer, line_layer])
lonboard_toc = lonboard.controls.make_TOC(lonboard_map)
display(lonboard_map)
display(lonboard_toc)
```

### Table of contents is reactive to changes

If you add a layer to the map after it is created, the table of content will dynamically refresh itself and show the new layer

```python
lonboard_map.add_layer(point_layer)
```
