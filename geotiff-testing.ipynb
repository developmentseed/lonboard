```python
from lonboard.experimental._surface import SurfaceLayer
```

```python
from lonboard import Map
```

```python
import rasterio

path = "/Users/kyle/github/developmentseed/deck.gl-raster/data/Annual_NLCD_LndCov_2024_CU_C1V1.tif"
src = rasterio.open(path)
```

```python
layer = SurfaceLayer.from_rasterio(src, out_shape=(1641, 2500), mesh_n_rows=1, mesh_n_cols=1, before_id="watername_ocean")
```

```python
from lonboard.basemap import MaplibreBasemap, CartoStyle
```

```python
basemap = MaplibreBasemap(mode="interleaved", style = CartoStyle.Positron)
```

```python
m = Map(layer, basemap=basemap, height=800)
```
